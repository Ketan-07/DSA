# Write a Program to input a single Linked List inside a stack data structure

* By Me :-

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class Stack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top is None

    def push(self, data):
        new_node = Node(data)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")
        popped_value = self.top.data
        self.top = self.top.next
        return popped_value

    def display(self):
        current = self.top
        while current is not None:
            print(current.data, end=" -> ")
            current = current.next
        print("None")


# Create a stack using linked list
stack = Stack()

# Input values into the stack
n = int(input("Enter number of elements: "))
for _ in range(n):
    value = int(input("Enter value: "))
    stack.push(value)

# Display the stack
print("Stack elements (top to bottom):")
stack.display()
stack.pop()
stack.display()
```

# Write a program to perform the following task of BookStack

* A person arranging his books , help him to arrange the books in the form of a stack data structure .
* Arrange the books on above another and if reached limited size of five then add the book in the secind stack automatically 
* So each stack contains five books and all the stacks is present in a outer stack, so inner stack has a size limit whereas outer stack has no size limit 
* Create a methods like push(),display(),pop(),pop_at()(means the stack number from where to remove)

```python
class Book_Stack:
    def __init__(self,maxSize):
        self.maxSize = maxSize
        self.s1 = []

    def push(self,val):
        if len(self.s1) > 0 and len(self.s1[-1]) < self.maxSize:
            self.s1[-1].append(val)
        else:
            self.s1.append([val])

    def display(self):
        if len(self.s1) != 0:
            print(self.s1)
        else:
            print("No elements")

    def pop(self):
        if(len(self.s1) == 0):
            return None
        elif len(self.s1) and len(self.s1[-1]) > 0:
            self.s1[-1].pop()
        else:
            self.s1.pop()

    def pop_at(self,stk_num):
        if len(self.s1)> 0:
            if stk_num <= len(self.s1):
                self.s1[stk_num-1].pop()
            else:
                print("No stcak is present at given num")
        else:
            print("Stack is empty")
        

    def peek(self): 
        if len(self.s1) > 0:
            print(self.s1[-1][-1])
        else:
            print('Stack is Empty')

    def peek_at(self,stk_num):
        if len(self.s1) > 0:
            if stk_num <= len(self.s1) and stk_num > 0:
                print(self.s1[stk_num-1][-1])
            else:
                print("No stack num")
        else:
            print("Stack has no elements")

    def push_at(self,stk_num,val):
        if stk_num <= len(self.s1):
            if len(self.s1[stk_num-1]) < self.maxSize :
                self.s1[stk_num-1].append(val)
            else:
                print("Stack rerached it's limit")
        else:
            print("No stack num")

b1 = Book_Stack(4)
b1.push("One")
b1.push("Two")
b1.push("Three")
b1.push("Four")
b1.push("Five")
b1.push("Six")
b1.push("Seven")
b1.push("Eight")
b1.push("Nine")
b1.push("Ten")
b1.push("Eleven")
b1.push("Twelve")
b1.display()
```

# Write a program to enqeue th elments to queue data structures and push the entire queue to stack data structure and remove the element by using rules of both stack and queue

```python
class stack:
    def __init__(self):
        self.s1 = []
    
    def push(self,val):
        self.s1.append(val)

    def display(self):
        for i in range(len(self.s1)-1,-1,-1):
            print(self.s1[i])

    def pop(self):
        if len(self.s1) == 0:
            return None
        return self.s1.pop()
    
class Queue_Stack:
    def __init__(self):
        self.stack1 = stack()
        self.stack2 = stack()

    def enqueue(self,val):
        self.stack1.push(val)

    def dequeue(self):
        while len(self.stack1.s1):
            self.stack2.push(self.stack1.pop())
        res = self.stack2.pop()
        while len(self.stack2.s1):
            self.stack1.push(self.stack2.pop())
        return res
    
    def display(self):
        self.stack1.display()

qs = Queue_Stack()
qs.enqueue("A")
qs.enqueue("B")
qs.enqueue("C")
qs.enqueue("D")
qs.enqueue("E")
qs.display()
print()
print("Removed Element :",qs.dequeue())
print()
qs.display()
```
