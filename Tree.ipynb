### Height of a tree

```python
class Tree:
    def __init__(self,data):
        self.prev = None
        self.data = data
        self.next = None

    def insert_the_element(self,val):
        if self.data:
            if val < self.data:
                if self.prev is None:
                    self.prev = Tree(val)
                else:
                    self.prev.insert_the_element(val)
            elif val > self.data:
                if self.next is None:
                    self.next = Tree(val)
                else:
                    self.next.insert_the_element(val)

    def display(self):
        if self.prev:
            self.prev.display()
        print(self.data,end=" -> ")
        if self.next:
            self.next.display()

t1 = Tree(15)
t1.insert_the_element(12)
t1.insert_the_element(20)
t1.insert_the_element(25)
t1.insert_the_element(18)
t1.insert_the_element(14)
t1.insert_the_element(9)
t1.insert_the_element(6)
t1.insert_the_element(10)
t1.display()
```
