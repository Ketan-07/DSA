# Single_Linked_List

```python
class Node:
    def __init__(self,data):
        self.data=data
        self.addr=None

class Single_Linked_list:
    def __init__(self):
        self.head=None
    
    def insert_at_last(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            temp = self.head
            while temp.addr != None:
                temp = temp.addr
            temp.addr = newNode
            
    def display(self):
        if self.head == None:
            print("No nodes to display")
        else:
            temp = self.head
            while temp:
                print(temp.data,end=" -> ")
                temp=temp.addr
            print()
    
    def length(self):
        if self.head == None:
            print("No nodes to count")
        else:
            temp=self.head
            cnt=0
            while temp:
                cnt+=1
                temp = temp.addr
            return cnt

    def insert_at_first(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            newNode.addr = self.head
            self.head = newNode

    def insert_at_loc(self,val,loc):
        newNode = Node(val)
        if loc <= 0 :
            print("Enter location above 0")
        elif loc==1:
            newNode.addr = self.head
            self.head = newNode
        elif loc == self.length() + 1:
            temp = self.head
            while temp.addr != None:
                temp = temp.addr
            temp.addr = newNode
        elif loc > self.length():
            print("Enter Location lessthan ->",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.addr != None and cnt < loc - 1 :
                temp = temp.addr
                cnt += 1
            newNode.addr = temp.addr
            temp.addr = newNode

    def delete_at_last(self):
        if self.head is None:
            print("No nodes to delete")
        elif self.length() == 1:
            self.head = None
        else:
            temp = self.head
            while temp.addr.addr != None:
                temp = temp.addr
            temp.addr = None

    def delete_at_first(self):
        if self.head is None:
            print("No nodes to remove")
        else:
            self.head = self.head.addr
        
    def delete_at_loc(self,loc):
        if loc <= 0 :
            print("Enter location above 0")
        elif loc==1:
            self.head = self.head.addr
        elif loc == self.length():
            temp = self.head
            while temp.addr.addr != None:
                temp = temp.addr
            temp.addr = None
        elif loc > self.length():
            print("Enter Location lessthan ->",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.addr != None and cnt < loc - 1 :
                temp = temp.addr
                cnt += 1
            temp.addr = temp.addr.addr

# sll = Single_Linked_list()
# sll.insert_at_last(10)
# sll.insert_at_last(20)
# sll.insert_at_last(30)
# sll.insert_at_last(40)
# sll.display()
# sll.insert_at_first(50)
# sll.display()
# print("No of Nodes in Linked List -",sll.length())
# sll.insert_at_loc(70,2)
# sll.display()
# print("No of Nodes in Linked List -",sll.length())
# sll.delete_at_last()
# sll.display()
# print("No of Nodes in Linked List -",sll.length())
# sll.delete_at_first()
# sll.display()
# print("No of Nodes in Linked List -",sll.length())
# sll.delete_at_loc(2)
# sll.display()
# print("No of Nodes in Linked List -",sll.length())
# sll.delete_at_loc(5)             # Invalid location, so returns error message

s1= Single_Linked_list()
while True:
    print('-----Single Linked List Operations-----')
    print('1. Insert at Last \n2. Insert at First \n3. Insert at Location \n4. Delete at Last \n5. Delete at First \n6. Delete at Location \n7. Display \n8. Length of Linked List \n9. Exit')
    print()
    option = int(input('Enter your choice: '))
    match option:
        case 1:
            val = input("Enter the value")
            s1.insert_at_last(val)
        case 2:
            val = input("Enter the value")
            s1.insert_at_first(val)
        case 3:
            val = input("Enter the value")
            loc = int(input("Enter the location"))
            s1.insert_at_loc(val,loc)
        case 4:
            s1.delete_at_last()
        case 5:
            s1.delete_at_first()
        case 6:
            loc = int(input("Enter the location"))
            s1.delete_at_loc(loc)
        case 7:
            s1.display()
        case 8:
            print("No of Nodes in Linked List -",s1.length())
        case 9:
            print("Exiting the program")
            break  
```

# Circular_Linked_List

```python
class node:
    def __init__(self, data):
        self.data = data
        self.addr = None
    
class Circular_Linked_List:
    def __init__(self):
        self.head = None

    def insert_at_last(self, val):
        newNode = node(val)
        if self.head is None:
            self.head = newNode
            self.head.addr = newNode
        else:
            temp = self.head
            while temp.addr != self.head:
                temp = temp.addr
            temp.addr = newNode
            newNode.addr = self.head

    def display(self):
        if self.head is None:
            print("No nodes to display")
        else:
            temp = self.head
            while temp:
                print(temp.data,end="->")
                temp = temp.addr
                if temp == self.head:
                    break
            print()

    def length(self):
        if self.head is None:
            print("No nodes to count")
        else:
            temp = self.head
            cnt = 0
            while temp:
                cnt += 1
                temp = temp.addr
                if temp == self.head:
                    break
            return cnt
    
    def insert_at_first(self,val):
        newNode = node(val)
        if self.head is None:
            self.head = newNode
            self.head.addr = newNode
        else:
            temp = self.head
            while temp.addr != self.head:
                temp = temp.addr
            temp.addr = newNode
            newNode.addr = self.head
            self.head = newNode

    def insert_at_loc(self, val, loc):
        newNode = node(val)
        if loc <= 0:
            print("Invalid location, enter above 0")
        elif loc == 1:
            self.insert_at_first(val)
        elif loc == self.length() + 1:
            self.insert_at_last(val)
        elif loc > self.length():
            print("Invalid location, enter below", self.length() + 1)
        else:
            temp = self.head
            cnt = 1
            while temp.addr != self.head and cnt < loc - 1:
                temp = temp.addr
                cnt += 1
            newNode.addr = temp.addr
            temp.addr = newNode
    
    def delete_at_last(self):
        if self.head is None:
            print("No nodes to remove")
        else:
            temp = self.head
            while temp.addr.addr != self.head:
                temp = temp.addr
            temp.addr = self.head

    def delete_at_first(self):
        if self.head is None:
            print("No nodes to remove")
        else:
            temp = self.head
            while temp.addr != self.head:
                temp = temp.addr
            temp.addr = self.head.addr
            self.head = self.head.addr

    def delete_at_loc(self, loc):
        if self.head is None:
            print("No nodes to remove")
        elif loc <= 0:
            print("Invalid location, enter above 0")
        elif loc == 1:
            self.delete_at_first()
        elif loc == self.length():
            self.delete_at_last()
        elif loc > self.length():
            print("Invalid location, enter below", self.length() + 1)
        else:
            temp = self.head
            cnt = 1
            while temp.addr != self.head and cnt < loc - 1:
                temp = temp.addr
                cnt += 1
            temp.addr = temp.addr.addr

c1 = Circular_Linked_List()
c1.insert_at_last('A')
c1.insert_at_last('N')
c1.insert_at_first('E')
c1.insert_at_first('K')
c1.insert_at_loc('T', 3)
c1.display()
c1.delete_at_last()
c1.display()
c1.delete_at_first()
c1.display()
c1.delete_at_loc(3)
c1.display()
```

# Double_Linked_List 

```python
class Node:
    def __init__(self,Data):
        self.prev = None
        self.data = Data
        self.next = None

class Double_Linked_list:
    def __init__(self):
        self.head = None

    def insert_at_last(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            temp = self.head
            while temp.next != None:
                temp = temp.next
            temp.next = newNode
            newNode.prev = temp

    def display(self):
        if self.head is None:
            print("No nodes to display")
        else:
            temp = self.head
            while temp:
                print(temp.data,end=" <=> ")
                temp=temp.next
            print()

    def length(self):
        if self.head is None:
            print("No nodes to count")
        else:
            temp = self.head
            cnt = 0
            while temp:
                cnt += 1
                temp = temp.next
            return cnt
        
    def insert_at_first(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
        else:
            newNode.next = self.head
            self.head.prev = newNode
            self.head = newNode

    def insert_at_loc(self,val,loc):
        newNode = Node(val)
        if loc <= 0:
            print("Enter loc above 0")
        elif loc == 1:
            self.insert_at_first(val)
        elif loc == self.length() + 1:
            self.insert_at_last(val)
        elif loc > self.length():
            print("Enter a location lessthan :",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.next != None and cnt < loc-1:
                temp = temp.next
                cnt += 1
            temp.next.prev = newNode
            newNode.next = temp.next
            newNode.prev = temp
            temp.next = newNode
            
    def delete_at_last(self):
        if self.head is None:
            print("No nodes to remove")
        elif self.length() == 1:
            self.head = None
        else:
            temp = self.head
            while temp.next.next != None:
                temp = temp.next
            temp.next.prev = None
            temp.next = None

    def delete_at_first(self):
        if self.head is None:
            print("No nodes to remove")
        elif self.length() == 1:
            self.head = None
        else:
            self.head.next.prev = None
            self.head = self.head.next

    def delete_at_loc(self,loc):
        if loc <= 0:
            print("Enter loc above 0")
        if loc == 1:
            self.delete_at_first()
        elif loc > self.length():
            print("Enter a location lessthan :",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.next != None and cnt < loc-1:
                temp = temp.next
                cnt += 1
            temp.next = temp.next.next
            temp.next.prev = temp 

    def search_ele(self,val):
        if self.head is None:
            print("No nodes")
        else:
            temp = self.head
            while temp.next != None:
                if temp.data == val:
                    print("Value found")
                    print("Previous value",temp.prev.data if temp.prev != None else "previous value is None")
                    print("Next value",temp.next.data if temp.next != None else "next value is None")
                    break
                temp = temp.next
            else:
                print("Value not found")

dll = Double_Linked_list()
dll.insert_at_last(10)
dll.insert_at_last(20)
dll.insert_at_last(30)
dll.insert_at_last(40)
dll.display()
print("Total nodes in dll",dll.length())
dll.insert_at_first(50)
dll.display()
print("Total nodes in dll",dll.length())
dll.insert_at_loc(60,3)
dll.display()
print("Total nodes in dll",dll.length())
dll.delete_at_first()
dll.display()
print("Total nodes in dll",dll.length())
dll.delete_at_last()
dll.display()
print("Total nodes in dll",dll.length())
dll.delete_at_loc(3)
dll.display()
print("Total nodes in dll",dll.length())
dll.insert_at_last(70)
dll.insert_at_last(80)
dll.insert_at_last(90)
dll.insert_at_last(100)
dll.display()
print("Total nodes in dll",dll.length())
dll.search_ele(80)
```

# Circular_Double_Linked_List 

```python
class Node:
    def __init__(self,data):
        self.prev = None
        self.data = data
        self.next = None

class Circular_Double_Linked_list:
    def __init__(self):
        self.head = None
    
    def insert_at_last(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
            newNode.prev = newNode
            newNode.next = newNode
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = newNode
            newNode.prev = temp
            newNode.next = self.head
            self.head.prev = newNode

    def display(self):
        if self.head is None:
            print("No nodes to display")
        else:
            temp = self.head
            while temp:
                print(temp.data,end=" -> ")
                temp = temp.next
                if temp == self.head:
                    break
            print()

    def length(self):
        if self.head is None:
            print("The length is 0")
        else:
            temp = self.head
            cnt = 0
            while temp:
                cnt += 1
                temp = temp.next
                if temp==self.head:
                    break
            return cnt

    def insert_at_first(self,val):
        newNode = Node(val)
        if self.head is None:
            self.head = newNode
            newNode.next = newNode
            newNode.prev = newNode
        else:
            temp = self.head
            while temp.next != self.head:
                temp = temp.next
            temp.next = newNode
            newNode.next = self.head
            newNode.prev = temp
            self.head.prev = newNode
            self.head = newNode

    def insert_at_loc(self,val,loc):
        newNode = Node(val)
        if loc <= 0:
            print("Enter a location above 0")
        elif loc == 1:
            self.insert_at_first(val)
        elif loc == self.length()+1:
            self.insert_at_last(val)
        elif loc > self.length():
            print("Enter a Location less than :",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.next != self.head and cnt < loc-1:
                temp = temp.next
                cnt += 1
            temp.next.prev = newNode
            newNode.next = temp.next
            newNode.prev = temp
            temp.next = newNode

    def delete_at_last(self):
        if self.head is None:
            print("No nodes to remove")
        elif self.length == 1:
            self.head = None
        else:
            temp = self.head
            while temp.next.next != self.head:
                temp = temp.next
            temp.next = self.head
            self.head.prev = temp

    def delete_at_first(self):
        if self.head is None:
            print("No nodes to remove")
        elif self.length == 1:
            self.head = None
        else:
            temp = self.head
            while temp.next != self.head:
                temp  = temp.next
            temp.next = self.head.next
            self.head.next.prev = temp
            self.head = self.head.next
        
    def delete_at_loc(self,loc):
        if loc <= 0:
            print("Enter location above 0")
        elif loc == 1:
            self.delete_at_first()
        elif loc == self.length():
            print("Enter location lessthan or equal to ",self.length())
        else:
            temp = self.head
            cnt = 1
            while temp.next != None and cnt < loc - 1:
                temp = temp.next
                cnt += 1
            temp.next = temp.next.next
            temp.next.prev = temp


cdll = Circular_Double_Linked_list()
cdll.insert_at_last('A')  
cdll.insert_at_last('B')  
cdll.insert_at_last('C')
cdll.display()
print("Total no of nodes in Circular_Double_Linked_List",cdll.length())
cdll.insert_at_first('D')
cdll.insert_at_first('F')
cdll.display()
print("Total no of nodes in Circular_Double_Linked_List",cdll.length())
cdll.insert_at_loc('E',3)
cdll.insert_at_loc('H',6)
cdll.display()
print("Total no of nodes in Circular_Double_Linked_List",cdll.length())
cdll.delete_at_last()
cdll.display()
print("Total no of nodes in Circular_Double_Linked_List",cdll.length())
cdll.delete_at_first()
cdll.display()
print("Total no of nodes in Circular_Double_Linked_List",cdll.length())
cdll.delete_at_loc(4)
cdll.display()
print("Total no of nodes in Circular_Double_Linked_List",cdll.length())
```
