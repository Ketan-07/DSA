# Eg 2 - Write a program to find the factorial of a given number without using loops

```python
def fact(n):
    if(n==1):
        return 1
    else:
        return n * fact(n-1)
n=int(input("Enter a number"))
print(f'Factorial of {n} = {fact(n)}')
```

# Eg 3 -Write a program to find the sum of first 10 natural numbers without for loop

```python
def sum(n):
    if(n==0):
        return 0
    else:
        return n + sum(n-1)
n=int(input("Enter a number"))
print(f'Sum of first {n} natural number  = {sum(n)}')
```

# Eg 4 -Sum of digits of given number using recurssion

```python
def sumOfdigit(n):
    if n==0:
        return 0
    else:
        return (n%10) + sumOfdigit(n//10)      
n=int(input("Enter a number"))
print(f'Sum of digits of number {n} = {sumOfdigit(n)}')      
```

# Write a program to find the sum of elements of list using recurssion

```python
def sum_of_list(l):
    if len(l)==0:
        return 0
    else:
        return l[0] + sum_of_list(l[1:])

print(sum_of_list([5,2,4,7,8,9,3,6]))
```

# Write a program to find the product of elements of list using recurssion

```python
def mul_of_list(l):
    if len(l)==0:
        return 1
    else:
        return l[0] * mul_of_list(l[1:])

print(mul_of_list([2,4,7,6]))
```

# Write a program to find the reverse of a string using recurssion

```python
def reverse_str(str):
    if len(str)==0:
        return ""
    else:
        return str[-1] + reverse_str(str[:-1])
str=input("Enter a String")
print(reverse_str(str))
```

# Check a given string is pailndrome or not

```python
def reverse_str(str):
    if len(str)==0:
        return ""
    else:
        return str[-1] + reverse_str(str[:-1])
str=input("Enter a String")
print("Is palindrome" if str == reverse_str(str) else "Not Palindrome")
```

```python
def isPalindrome(s1):
    if len(s1)==0:
        return "String is palindrome"
    elif s1[0] != s1[-1]:
        return "String is not Palindrome"
    else:
        return isPalindrome(s1[1:-1])

print(isPalindrome("MADAM"))
print(isPalindrome("Feku"))
print(isPalindrome("MOM"))
```
