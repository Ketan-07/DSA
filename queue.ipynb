# Linear Queue

```python
class queue:
    def __init__(self,maxSize):
        self.q1=[]
        self.maxSize=maxSize
    
    def isFull(self):
        if len(self.q1) == self.maxSize:
            return True
        else:
            return False
    
    def isEmpty(self):
        if len(self.q1)==0:
            return True
        else:
            return False
        
    def enqueue(self,val):
        if self.isFull():
            print("Queue is full")
        else:
            self.q1.append(val)
        
    def peek(self):
        if self.isEmpty():
            print("No elements to peek")
        else:
            print("Peek Element is :",{self.q1[0]})

    def display(self):
        if self.isEmpty():
            print("No Elements to display")
        else:
            print(self.q1)

    def dequeue(self):
        if self.isEmpty():
            print("No elements to delete")
        else:
            self.q1.pop(0)

q=queue(5)

q.enqueue("A")
q.enqueue("B")
q.enqueue("C")
q.display()
q.peek()
q.dequeue()
q.display()
q.peek()
```

# Circular Queue

```python
class Circular_Queue:
    def __init__(self,maxSize):
        self.maxSize=maxSize
        self.c1=[None]*maxSize
        self.front = -1
        self.rear = -1
    
    def isFull(self):
        if self.front == 0 and self.rear + 1 == self.maxSize:
            return True
        elif self.rear + 1 == self.front:
            return True
        else:
            False

    def isEmpty(self):
        if self.front == -1 and self.rear + 1 == -1:
            return True
        else:
            return False

    def display(self):
        print(self.c1)
        print("front ->",self.front)
        print("rear ->",self.rear)

    def enqueue(self,val):
        if self.isFull():
            print("Circular Queue is full")
        else:
            if self.rear + 1 == self.maxSize:
                self.rear = 0
            else:
                self.rear += 1
                if self.front == -1:
                    self.front = 0
            
            self.c1[self.rear] = val

    def dequeue(self):
        if self.isEmpty():
            pass
        else:
            temp = self.front
            if self.front == self.rear :
                self.front = -1
                self.rear = -1
            elif self.front + 1 == self.maxSize:
                self.front= 0
            else:
                self.front += 1
            self.c1[temp] = None

        
    def peek_ele(self):
        if self.isEmpty():
            print("No elements to Circular Queue")
        else:
            print(self.c1[self.front])

cq=Circular_Queue(5)
# cq.display()
cq.enqueue(10)
cq.enqueue(20)
cq.enqueue(30)
cq.enqueue(40)
cq.enqueue(50)
cq.display()
cq.enqueue(60)
cq.display()
cq.peek_ele()
cq.dequeue()
cq.dequeue()
cq.display()
cq.peek_ele()
cq.enqueue(60)
cq.enqueue(70)
cq.display()
cq.peek_ele()
```
