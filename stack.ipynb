 # Stack without limited size

```python
class Stack:
    def __init__(self):
        self.s1=[]

    def display(self):
            if len(self.s1)!=0:
                for i in range(len(self.s1)-1,-1,-1):
                    print(self.s1[i] , end=" ")
                print()
            else:
                print("No elements to display")

    def push(self,val):
            self.s1.append(val)

    def pop(self):
            self.s1.pop()

    def peek_ele(self):
            print("Top most element in stack :",self.s1[-1])

st=Stack()
st.push(10)
st.push(20)
st.push(30)
st.push(40)
st.push(50)
st.push(60)
st.push(70)
st.display()
st.pop()
st.pop()
st.display()
st.push(80)
st.display()
st.peek_ele()
```

# Stack with Limited Size

```python
class Stack:
    
    def __init__(self,maxSize):
        self.s1=[]
        self.maxSize=maxSize

    def isFull(self):
          if len(self.s1)==self.maxSize:
                return True
          else:
                return False
          
    def isEmpty(self):
          if len(self.s1) == 0:
                return True
          else:
                return False

    def display(self):
            if self.isEmpty():
                  print("No Element in Stack")
            else:
                for i in range(len(self.s1)-1,-1,-1):
                    print(self.s1[i] , end=" ")
                print()
            
    def push(self,val):
            if self.isFull():
                  print("Stack is Full")
            else:
                self.s1.append(val)

    def pop(self):
            if self.isEmpty():
                  print("Stack has no Elements")
            else:
                self.s1.pop()

    def peek_ele(self):
            print("Top most element in stack :",self.s1[-1])

st=Stack(5)
st.display()
st.push(10)
st.push(20)
st.peek_ele()
st.push(30)
st.display()
st.push(40)
st.push(50)
st.push(60)
st.display()
```

# Write a program using stack to check the given parenthesis is balanced or not

```python
def balanced_parenthesis(s1):
    stack=[]
    values={'}':'{',')':'(',']':'['}

    for char in s1:
        if char in "({[":
            stack.append(char)
        elif char in '})]':
            if not stack or stack.pop() != values[char]:
                return False
        return len(stack) == 0


# s1="[](){}"
s1="{(}"
if balanced_parenthesis(s1):
    print("Balanced")
else:
    print("Not Balanced")
```
